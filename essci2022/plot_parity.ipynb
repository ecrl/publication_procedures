```python
from csv import DictReader

with open('pred_results.csv', 'r') as csv_file:
    reader = DictReader(csv_file)
    rows = [r for r in reader]
csv_file.close()
```

```python
from matplotlib import pyplot as plt

actual = [float(r['actual']) for r in rows]
pred = [float(r['pred']) for r in rows]

plt.scatter(actual, pred, color='blue')
plt.xlabel('Experimental CN')
plt.ylabel('Predicted CN')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, median_absolute_error
from math import sqrt

print(f'Median absolute error: {median_absolute_error(actual, pred)}')
print(f'RMSE: {sqrt(mean_squared_error(actual, pred))}')
```
