```python
from tokenizer import SMILESTokenizer

smi = SMILESTokenizer()
smi.fit([
    'CC(C)(C)O',
    'COC1=C(C(=CC=C1)OC)O',
    'CC(CO)O',
    'CCCC(=O)OCC(COC(=O)CCC)OC(=O)CCC',
    'C1=CC=C(C=C1)N',
    'CCCCCC=O',
    'C1=C(OC(=C1)C=O)CO',
    'CC(C)O',
    'CN(C)C=O',
    'CCCO'
], n_merges=20)
print(smi._vocab)
```
