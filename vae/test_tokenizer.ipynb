```python
from preprocessing import SMILESEncoder, SMILESTokenizer

smiles = [
    'CC(C)(C)O',
    'COC1=C(C(=CC=C1)OC)O',
    'CC(CO)O',
    'CCCC(=O)OCC(COC(=O)CCC)OC(=O)CCC',
    'C1=CC=C(C=C1)N',
    'CCCCCC=O',
    'C1=C(OC(=C1)C=O)CO',
    'CC(C)O',
    'CN(C)C=O',
    'CCCO'
]

tokenizer = SMILESTokenizer(smiles, n_merges=16)
tokens = tokenizer.tokenize(smiles)
print(tokens)
print()
print(tokenizer.vocab)
```

```python
encoder = SMILESEncoder(smiles, n_merges=16)
results = encoder.encode(smiles)
print(results.shape)
print()
print(results[0])
```

```python
print(encoder.decode(results))
```
